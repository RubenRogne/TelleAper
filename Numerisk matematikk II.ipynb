# Oppgave d)

Forskjellen mellom definisjoen av den deriverte og newtons kvotient er at du i newtons kvotient ikke lar Δx gå mot 0, men et lite tall. Dermed får vi et numerisk svar.  

```python
def deriverte(f,x, h = 1E-8):
    return (f(x+h)-f(x)) / h

def f(x):
    return x**2 + 4*x

deriverte(f,2)
```

# Oppgave g)

```python
from pylab import *

tid = []
fart = []
aks = []


fil = open("bevegelse.txt","r")

for linje in fil:
    data = linje.split(",")
    tid.append(float(data[0]))
    fart.append(float(data[1]))
    
fil.close()



for i in range(0,len(fart)):
    aks.append((fart[i]-fart[i-1])/(tid[i]-tid[i-1]))



x1 = tid
y1 = fart
x2 = tid
y2 = aks
plot(x1,y1)
plot(x2,y2)
grid("on")
show()
```

# Oppgave i)

```python
from pylab import *

fil = imread("bilde.png")



dfdx = abs(fil[1:-1,2:,:] - fil[1:-1,1:-1,:])
imshow(3*dfdx)
savefig("dfdx.png")


dfdy = abs(fil[2:,1:-1,:]- fil[1:-1,1:-1,:])
imshow(3*dfdy)
savefig("dfdy.png")


grad = sqrt(dfdx**2 + dfdy**2)
imshow(3*grad)
savefig(grad.png)
```
